```python
fs_real = {'0.0': root_open('/home/christian/msc/results/131211/golden_05-1_1-2_ignore_no_trig/output_real.root', 'read'),
           '2.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_2/output_real.root', 'read'),
           '3.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_3/output_real.root', 'read'),
           '4.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_4/output_real.root', 'read'),
           '5.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_5/output_real.root', 'read')}
fs_reco = {'0.0': root_open('/home/christian/msc/results/131211/golden_05-1_1-2_ignore_no_trig/output_reconstructed.root', 'read'),
           '2.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_2/output_reconstructed.root', 'read'),
           '3.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_3/output_reconstructed.root', 'read'),
           '4.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_4/output_reconstructed.root', 'read'),
           '5.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_5/output_reconstructed.root', 'read')}

from collections import OrderedDict
from ROOT import Double
from copy import deepcopy
err = Double()
yields_peak = []
yields_away = []
yields_near = []
ofs_real = OrderedDict(sorted(fs_real.items()))

plt.close('all')
figsize= (2.95 /2.5 * 5, 2.95/4*3)
fig, axs, = plt.subplots(1, 5, sharex=True, sharey=True, figsize=figsize, )
lns = []
for t, f_real, color, ax in zip(ofs_real.keys(), ofs_real.values(), colors, axs):
    h_reco = fs_reco.get(t).processed.subtracted_corrected.Get('Subtraction_thresh_'+t)
    h_real = f_real.processed.subtracted_corrected.Get('Subtraction_thresh_'+t)
    min_bin = h_real.projection_x().FindBin(1.3)
    zyam_real = h_real.projection_x().Integral(min_bin-1, min_bin+1) / (3 * 32)
    zyam_reco = h_reco.projection_x().Integral(min_bin-1, min_bin+1) / (3 * 32)
    scale = zyam_real / zyam_reco
    
    h_real.GetXaxis().SetRange(1, 17)
    h_real_py = asrootpy(h_real.projection_y()) / 17
    h_real_py.color = colors[0]
    lns.append(rplt.errorbar(h_real_py, axes=ax, label='Data', markersize=3))
    
    h_reco.GetXaxis().SetRange(1, 17)
    h_reco_py = asrootpy(h_reco.projection_y())* scale / 17
    h_reco_py.color = colors[1]
    lns.append(rplt.errorbar(h_reco_py, axes=ax, label='MC recon.', markersize=3))
    
    lower = h_reco_py.FindBin(-.8 + .01)
    upper = h_reco_py.FindBin(.8- .01)

    x = list(h_reco_py.x())[lower-1:upper]  # index != bin
    y1 = list(h_reco_py.y())[lower-1:upper]
    y2 = list(h_real_py.y())[lower-1:upper]
    ax.fill_between(x, y1, y2, color='grey', alpha='0.5')
   
    # using projection onto dphi for yield extraction
    # peak region
    y_range = (h_real_py.FindBin(-.8 + .01), h_real_py.FindBin(.8- .01))
    h_real.GetYaxis().SetRange(*y_range)
    h_real_px = asrootpy(h_real.projection_x()) / (y_range[1] - y_range[0] +1)
    h_reco.GetYaxis().SetRange(*y_range)
    h_reco_px = asrootpy(h_reco.projection_x()) / (y_range[1] - y_range[0] +1)
    diff_histo = h_real_px - h_reco_px * scale
    lower = diff_histo.FindBin(-pi/2 + .01)
    upper = diff_histo.FindBin(pi/2 - 0.01)
    yields_peak.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))
    
    # near side
    h_real.GetYaxis().SetRange(0,33); h_reco.GetYaxis().SetRange(0,33)
    diff_histo = extract_yield(h_real)[0] - extract_yield(h_reco)[0] * scale
    lower = diff_histo.FindBin(-pi/2 + .01)
    upper = diff_histo.FindBin(pi/2 - 0.01)
    yields_near.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))
    
    # away side
    h_real.GetYaxis().SetRange(0,33); h_reco.GetYaxis().SetRange(0,33)
    lower = diff_histo.FindBin(pi/2 + .01)
    upper = diff_histo.FindBin(3* pi/2 - 0.01)
    yields_away.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))

    plt.ylim(1.27, 1.45)
    plt.minorticks_on()
    ax.xaxis.major.locator.set_params(nbins=7)
    ax.set_xlabel(r'$\Delta\eta$')
    ax.set_title('Threshold \n$\SI{' + t + '}{GeV}$')
axs[0].legend(lns[0:2], [lab.get_label() for lab in lns[0:2]], loc=0)
axs[0].set_ylabel(r'$\frac{1}{N_{tr}}\frac{dN_{as}}{d\Delta \varphi}$'
                  '[rad$^{-1}$]')
plt.tight_layout(w_pad=.5)
fig.savefig('nearside_peak_yield.pdf')
```

```python
fs_real = {'0.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_0/output_real.root', 'read'),
           #'2.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_2/output_real.root', 'read'),
           '3.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_3/output_real.root', 'read'),
           '4.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_4/output_real_no_st.root', 'read'),
           #'5.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_5/output_real.root', 'read')
           }
fs_reco = {'0.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_0/output_reconstructed.root', 'read'),
           #'2.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_2/output_reconstructed.root', 'read'),
           '3.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_3/output_reconstructed.root', 'read'),
           '4.0': root_open('/home/christian/msc/results/finals/golden_1-2_2-4_thresh_4/output_reconstructed.root', 'read'),
           #'5.0': root_open('/home/christian/msc/results/finals/golden_05-1_1-2_thresh_5/output_reconstructed.root', 'read')
           }

from collections import OrderedDict
from ROOT import Double
from copy import deepcopy
err = Double()
yields_peak = []
yields_away = []
yields_near = []
ofs_real = OrderedDict(sorted(fs_real.items()))

plt.close('all')
figsize= (2.95 /2.5 * 5, 2.95/4*3)
fig, axs, = plt.subplots(1, 5, sharex=True, sharey=True, figsize=figsize, )
lns = []
for t, f_real, color, ax in zip(ofs_real.keys(), ofs_real.values(), colors, axs):
    h_reco = fs_reco.get(t).processed.subtracted_corrected.Get('Subtraction_thresh_'+t)
    h_real = f_real.processed.subtracted_corrected.Get('Subtraction_thresh_'+t)
    min_bin = h_real.projection_x().FindBin(1.3)
    zyam_real = h_real.projection_x().Integral(min_bin-1, min_bin+1) / (3 * 32)
    zyam_reco = h_reco.projection_x().Integral(min_bin-1, min_bin+1) / (3 * 32)
    scale = zyam_real / zyam_reco
    
    h_real.GetXaxis().SetRange(1, 17)
    h_real_py = asrootpy(h_real.projection_y()) / 17
    h_real_py.color = colors[0]
    lns.append(rplt.errorbar(h_real_py, axes=ax, label='Data', markersize=3))
    
    h_reco.GetXaxis().SetRange(1, 17)
    h_reco_py = asrootpy(h_reco.projection_y())* scale / 17
    h_reco_py.color = colors[1]
    lns.append(rplt.errorbar(h_reco_py, axes=ax, label='MC recon.', markersize=3))
    
    lower = h_reco_py.FindBin(-.8 + .01)
    upper = h_reco_py.FindBin(.8- .01)

    x = list(h_reco_py.x())[lower-1:upper]  # index != bin
    y1 = list(h_reco_py.y())[lower-1:upper]
    y2 = list(h_real_py.y())[lower-1:upper]
    ax.fill_between(x, y1, y2, color='grey', alpha='0.5')
   
    # using projection onto dphi for yield extraction
    # peak region
    y_range = (h_real_py.FindBin(-.8 + .01), h_real_py.FindBin(.8- .01))
    h_real.GetYaxis().SetRange(*y_range)
    h_real_px = asrootpy(h_real.projection_x()) / (y_range[1] - y_range[0] +1)
    h_reco.GetYaxis().SetRange(*y_range)
    h_reco_px = asrootpy(h_reco.projection_x()) / (y_range[1] - y_range[0] +1)
    diff_histo = h_real_px - h_reco_px * scale
    lower = diff_histo.FindBin(-pi/2 + .01)
    upper = diff_histo.FindBin(pi/2 - 0.01)
    yields_peak.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))
    
    # near side
    h_real.GetYaxis().SetRange(0,33); h_reco.GetYaxis().SetRange(0,33)
    diff_histo = extract_yield(h_real)[0] - extract_yield(h_reco)[0] * scale
    lower = diff_histo.FindBin(-pi/2 + .01)
    upper = diff_histo.FindBin(pi/2 - 0.01)
    yields_near.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))
    
    # away side
    h_real.GetYaxis().SetRange(0,33); h_reco.GetYaxis().SetRange(0,33)
    lower = diff_histo.FindBin(pi/2 + .01)
    upper = diff_histo.FindBin(3* pi/2 - 0.01)
    yields_away.append((diff_histo.IntegralAndError(lower, upper, err, 'width'), deepcopy(err)))

    plt.ylim(1.27, 1.45)
    plt.minorticks_on()
    ax.xaxis.major.locator.set_params(nbins=7)
    ax.set_xlabel(r'$\Delta\eta$')
    ax.set_title('Threshold \n$\SI{' + t + '}{GeV}$')
axs[0].legend(lns[0:2], [lab.get_label() for lab in lns[0:2]], loc=0)
axs[0].set_ylabel(r'$\frac{1}{N_{tr}}\frac{dN_{as}}{d\Delta \varphi}$'
                  '[rad$^{-1}$]')
plt.tight_layout(w_pad=.5)
#fig.savefig('nearside_peak_yield.pdf')
```
